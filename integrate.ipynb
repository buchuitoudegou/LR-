```python
import pandas as pd
import numpy as np
```

```python
t1 = pd.read_csv('result_lr.csv')
t2 = pd.read_csv('rf_result.csv')
t3 = pd.read_csv('nn_5_fold.csv')

t1 = t1['predicted']
t2 = t2['Predicted']
t3 = t3['Predicted']
```

```python
t1 = np.array(t1)
t2 = np.array(t2)
t3 = np.array(t3)
```

```python
result = []
for i in range(t1.shape[0]):
    tmp = t1[i] + t2[i] + t3[i]
    if tmp >= 2:
        result.append(1)
    else:
        result.append(0)
result
```

```python
output = pd.DataFrame(data={'id': list(range(1, len(result)+1)), 'Predicted': result})
output.to_csv('integrate.csv', index=False)
```
